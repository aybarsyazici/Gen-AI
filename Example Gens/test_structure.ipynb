```python
# We have 6 prompting methods. Each method is tested with GPT3.5 and GPT4.
# Thus we have 12 results in total.
import pandas as pd
results_map = {
    #"few_shot_gpt3_results.csv": 1,
    "few_shot_gpt4_results.csv": 2,
    #"one_shot_gpt3_results.csv": 3,
    "one_shot_gpt4_results.csv": 4,
    #"one_shot_first_try_gpt3_results.csv": 5,
    #"one_shot_first_try_gpt4_results.csv": 6,
    #"one_shot_short_gpt3_results.csv": 7,
    "one_shot_short_gpt4_results.csv": 8,
    #"one_shot_short_extra_info_gpt3_results.csv": 9,
    "one_shot_short_extra_info_gpt4_results.csv": 10,
    #"one_shot_extra_info_gpt3_results.csv": 11,
    "one_shot_extra_info_gpt4_results.csv": 12, # add ingredient rules
}

# Now for each of these results, load the csv and add an extra column that has the result number
# Append all the results together and create a final dataframe

final_df = pd.DataFrame()
for file_name, result_number in results_map.items():
    df = pd.read_csv(f'../dataset/{file_name}')
    df["result_number"] = result_number
    # pandas concat
    final_df = pd.concat([final_df, df])
```

```python
final_df.reset_index(drop=False, inplace=True)
```

```python
# Rename level_0 to original_recipe_number
final_df.rename(columns={"level_0": "original_recipe_number"}, inplace=True)
```

```python
final_df['row_number'] = final_df.index + 1
```

```python
groups = {}
# Groups variable will contain a dictionary,
# Where the key is the group number and the value is a list of tuples
# Where each tuple is a pair of rows that belong to the same group
# (row1,row2)
# i.e. groups = {1: [(row1,row2), (row3,row4), ...], 2: [(row5,row6), (row7,row8), ...], ...}

# group by index
for _, group in final_df.groupby("original_recipe_number"):
    # Get ALL of the rows that belong to this group
    rows = final_df[final_df["index"] == group["index"].iloc[0]]
    # Now shuffle and put them in groups of two, without replacement
    rows = rows.sample(frac=1).reset_index(drop=True)
    #grouped_rows = list(zip(rows.iloc[::2].values, rows.iloc[1::2].values))
    groups[_] = list(rows.values)
```
